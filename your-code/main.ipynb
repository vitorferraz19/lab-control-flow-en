## Words

```python
words = ['play', 'filling', 'bar', 'theatre', 'easygoing', 'date', 'lead', 'that', 'story',  'island']
```

**Print every word in upper case**

```python
for single_word in words:
    print(single_word.upper())
```

**Create a new list containing only words with 5 or more letters**

```python
words_five_more_chars = []
for single_word in words:
    if len(single_word) >= 5:
        words_five_more_chars.append(single_word)
print(words_five_more_chars)
```

**Print the first word starting with "t"**

```python
for single_word in words:
    if single_word[0] == "t":
        print(single_word)
        break                         #to force exiting the loop, otherwise, it'll print all words starting with a "t"
```
