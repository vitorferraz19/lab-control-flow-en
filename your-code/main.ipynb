## Words

```python
words = ['play', 'filling', 'bar', 'theatre', 'easygoing', 'date', 'lead', 'that', 'story',  'island']
```

**Print every word in upper case**

```python
for single_word in words:
    print(single_word.upper())
```

**Create a new list containing only words with 5 or more letters**

```python
words_five_more_chars = []
for single_word in words:
    if len(single_word) >= 5:                     #condition to check length higher or equal than 5
        words_five_more_chars.append(single_word)
print(words_five_more_chars)
```

**Print the first word starting with "t"**

```python
for single_word in words:
    if single_word[0] == "t":
        print(single_word)
        break                         #to force exiting the loop, otherwise, it'll print all words starting with a "t"
```

## Numbers

**Create a list containing the square of every number from 1 to 10**

```python
list_squares = []
for number in range(1,11):      #generates numbers from 1 to 10
    list_squares.append(number**2)       #squaring and appending 

print(list_squares)
```

**Print a list containing the square of every odd number from 1 to 10**

```python
list_squares_odd_numbers = []
for number in range(1,11):
    if number % 2 != 0:         # meaning, an odd number
        list_squares_odd_numbers.append(number**2)     #squaring and appending 
print(list_squares_odd_numbers)
```

**Create a list with the squares of all multiples of 8 below 1000**

```python
list_squares_multiples = []

for number in range(1,1000):
    if number%8 ==0:         # meaning, divisible by 8
        list_squares_multiples.append(number**2)       #squaring and appending 
print(list_squares_multiples)
```
